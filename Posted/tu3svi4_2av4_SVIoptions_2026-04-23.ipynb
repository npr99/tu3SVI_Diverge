# Obtain Data


## 1.1 Obtain data from about each variable in SVI data.

```python
# set data year
data_year = '2020'
geo_year = '2020' # geo year has to be 2010 or 2020

# dictionary of files to read in
SVI_files = {'SVInsight' : {'folderpath' : '../SourceData/UT_Preisser_SVI/',
                      'Tract' : f'tu3svi2_2av3_SetupUT_2025-04-08_Tract{data_year}.csv',
                      'BG'    : f'tu3svi2_2av3_SetupUT_2025-04-08_BG{data_year}.csv'},
             'CDC' : {'folderpath' : '../SourceData/www_atsdr_cdc_gov_placeandhealth_svi/',
                      'Tract' : f'tu3svi2_2bv2_SetupCDC_2024-09-06_Tract{data_year}.csv',
                      'BG'    : 'not available'},
             'HRRC': {'folderpath' : '../SourceData/texasatlas_arch_tamu_edu/',
                        'Tract' : f'tu3svi2_2cv3_SetupHRRC_2025-04-08_TractTX{data_year}.csv',
                        'BG'    : f'tu3svi2_2cv3_SetupHRRC_2025-04-08_BGTX{data_year}.csv'}
            }
```

```python
# start empty dictionary to store dataframes
Tract_dfs = {}
BG_dfs    = {}
# loop over each SVI file
for SVI_file in SVI_files:
    print("Reading in SVI files for :",SVI_file)
    # Get file information
    folderpath = SVI_files[SVI_file]['folderpath']
    Tract_file = SVI_files[SVI_file]['Tract']
    BG_file    = SVI_files[SVI_file]['BG']
    # Read in Tract file
    if Tract_file != 'not available':
        Tract_dfs[SVI_file] = pd.read_csv(folderpath + Tract_file, dtype={f'TRACT{geo_year}': str})
    else:
        print('No Tract file available')
    # Read in BG file
    if BG_file != 'not available':
        BG_dfs[SVI_file] = pd.read_csv(folderpath + BG_file, dtype={f'BG{geo_year}': str})
    else:
        print('No block group file available')
    
```

```python
### Merge dataframes
# loop through data frames in Tract_dfs and merge together on TRACT2010
print("Merging Tract dataframes")
Tract_df = pd.DataFrame()
for SVI_file in Tract_dfs:
    print("Merging :",SVI_file)
    if Tract_df.empty:
        Tract_df = Tract_dfs[SVI_file]
    else:
        Tract_df = Tract_df.merge(Tract_dfs[SVI_file], on=f'TRACT{data_year}', how='outer')
```

```python
# loop through data frames in Block group and merge together on BG2010
print("Merging Block Group dataframes")
BG_df = pd.DataFrame()
for SVI_file in BG_dfs:
    print("Merging :",SVI_file)
    if BG_df.empty:
        BG_df = BG_dfs[SVI_file]
    else:
        BG_df = BG_df.merge(BG_dfs[SVI_file], on=f'BG{data_year}', how='outer')
```

```python
# convert SVI Scores to Percent (0-100)
Tract_df[f'C{data_year}90012'] = Tract_df[f'C{data_year}90012']*100
Tract_df[f'H{data_year}90012'] = Tract_df[f'H{data_year}90012']*100
Tract_df[f'U{data_year}90012'] = Tract_df[f'U{data_year}90012']*100

# convert SVI Scores to Percent (0-100)
BG_df[f'H{data_year}90012'] = BG_df[f'H{data_year}90012']*100
BG_df[f'U{data_year}90012'] = BG_df[f'U{data_year}90012']*100

# Create TRACT ID from BG ID by removing the last digit
BG_df[f'TRACT{geo_year}'] = BG_df[f'BG{geo_year}'].str[:-1]
```

### Add Area Land and Water

```python
### Read in 2020 tract shapefile
sourcefolder = "../SourceData/www2_census_gov/tl_2020_48_tract"
sourcefile = "tl_2020_48_tract.shp"
file_path = os.path.join(sourcefolder, sourcefile)
tx_2020_tract_gdf = gpd.read_file(file_path)
# keep geoid and geometry
tx_2020_tract_gdf = tx_2020_tract_gdf[[f'GEOID', 'geometry','ALAND','AWATER']].copy(deep=True)
# rename geoid to BG2020
tx_2020_tract_gdf.rename(columns={'GEOID': f'TRACT{geo_year}'}, inplace=True)
# convert geoid to string
tx_2020_tract_gdf[f'TRACT{geo_year}'] = tx_2020_tract_gdf[f'TRACT{geo_year}'].astype(str)
# merge tract and SVI data
Tract_gdf = pd.merge(tx_2020_tract_gdf, Tract_df, on=f'TRACT{geo_year}', how='left')
Tract_gdf.head()
```

```python
### Read in 2020 BG shapefile
sourcefolder = "../SourceData/www2_census_gov/tl_2020_48_bg"
sourcefile = "tl_2020_48_bg.shp"
file_path = os.path.join(sourcefolder, sourcefile)
tx_2020_bg_gdf = gpd.read_file(file_path)
# keep geoid and geometry
tx_2020_bg_gdf = tx_2020_bg_gdf[[f'GEOID', 'geometry','ALAND','AWATER']].copy(deep=True)
# rename geoid to BG2020
tx_2020_bg_gdf.rename(columns={'GEOID': f'BG{geo_year}'}, inplace=True)
# convert geoid to string
tx_2020_bg_gdf[f'BG{geo_year}'] = tx_2020_bg_gdf[f'BG{geo_year}'].astype(str)
# merge bg and SVI data
BG_gdf = pd.merge(tx_2020_bg_gdf, BG_df, on=f'BG{geo_year}', how='left')
BG_gdf.head()
```

## Generate Codebook
https://github.com/npr99/pypdfcodebook

```python
from pypdfcodebook.pdfcb_03c_codebook import codebook
```

```python
from importlib.util import module_from_spec, spec_from_file_location
from pathlib import Path

filename = "tu3svi2_0bv1_SVIdatadictionary_2026-04-23_datastructure.py"
file_path = Path("../SourceData/") / filename

# Run .py code to get the base data structure.
spec = spec_from_file_location("svi_ds", file_path)
svi_ds = module_from_spec(spec)
spec.loader.exec_module(svi_ds)

# Copy and align data structure to dataframe columns.
datastructure = dict(svi_ds.DATA_STRUCTURE)

datastructure_bg, compare_df = svi_ds.match_data_structure_to_df(
    datastructure=datastructure,
    df=BG_df
    )

# Copy and align data structure to dataframe columns.
datastructure = dict(svi_ds.DATA_STRUCTURE)

datastructure_tract, compare_df = svi_ds.match_data_structure_to_df(
    datastructure=datastructure,
    df=Tract_df
    )
```

```python
codebook_output_folders = {"top": programname}
```

```python
cb = codebook(
    input_df=BG_df,
    header_title="SVI Analysis: Block Group Level Data",
    datastructure=datastructure_bg,
    projectoverview="../README.md",
    keyterms='',
    output_filename=f"{programname}_all_vars_bg2020_codebook",
    outputfolders=codebook_output_folders,
 )

# Generate the PDF codebook
cb.create_codebook()
```

```python
cb = codebook(
    input_df=Tract_df,
    header_title="SVI Analysis: Tract Level Data",
    datastructure=datastructure_tract,
    projectoverview="../README.md",
    keyterms='',
    output_filename=f"{programname}_all_vars_tract2020_codebook",
    outputfolders=codebook_output_folders,
 )

# Generate the PDF codebook
cb.create_codebook()
```
