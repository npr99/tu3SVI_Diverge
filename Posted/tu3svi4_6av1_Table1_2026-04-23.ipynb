## Clean Data
Need to keep the SVI, gencat, comcat, ACStable1v2v2

```python
# keep columns that are needed
keep_cols = ['SVI', 'year','gencatcode','comcatcode','gencat', 'comcat', 'label', 'ACStable1v2','newvarname']
table1_df = varmetadata_df[keep_cols].copy(deep=True)
# keep if gencatcode equal to or greater than 200 and less than 900
condition1 = table1_df['gencatcode'] >= 100
condition2 = table1_df['gencatcode'] < 900
table1_df = table1_df[condition1 & condition2]
```

```python
table1_df.head()
```

```python
# if comcat is missing replace with the label
condition = table1_df['comcat'].isnull()
table1_df.loc[condition, 'comcat'] = table1_df.loc[condition, 'label']
# Drop 'label' and 'gencatcode' columns
table1_dfv2 = table1_df.drop(columns=['label'])

table1_dfv2.head(20)
```

```python
# Using pivot_table to avoid errors on duplicate 'ACStable1v2' and 'SVI' entries
table1_dfv3 = table1_dfv2.pivot_table(index=['year','gencat','gencatcode','comcatcode','comcat'], columns='SVI', 
                                  values=['ACStable1v2','newvarname'], aggfunc='first')
# Reset the index
table1_dfv3.reset_index(inplace=True)
# sort by gencatcode and comcatcode
table1_dfv3.sort_values(by=['gencatcode','comcatcode'], inplace=True)
# replace missing with blank ''
table1_dfv3.fillna('', inplace=True)
# Display the first few rows of the transformed DataFrame
table1_dfv3.head()
```

```python
# rename columns and reset so only one level
table1_dfv3.columns = [' '.join(col).strip() for col in table1_dfv3.columns.values]
table1_dfv3.reset_index(drop=True, inplace=True)
table1_dfv3.head()
```

```python
# list columns
table1_dfv3.columns
```

### Fix some known issues
- HRRC does not actually use Median Home Value in SVI.
- HRRC does not actually use Rent Burden in SVI.
- HRRC uses table 25033 for renter households

```python
# for HRRC Median Home Value replace B25077 with ''
condition = table1_dfv3['ACStable1v2 HRRC'] == 'B25077'
table1_dfv3.loc[condition, 'ACStable1v2 HRRC'] = ''

# for rent burden replace B25070 with ''
condition = table1_dfv3['ACStable1v2 HRRC'] == 'B25070'
table1_dfv3.loc[condition, 'ACStable1v2 HRRC'] = ''
```

```python
table1_dfv3.head()
```

### Add extra rows

```python
# raname columns
table1_dfv3.rename(columns={'gencat':'General Category',
                            'comcat':'Common Category',
                            'ACStable1v2 CDC':'CDC',
                            'ACStable1v2 HRRC':'HRRC',
                            'ACStable1v2 UTP':'SVInsight',
                            'match':'Match'}, inplace=True)
```
